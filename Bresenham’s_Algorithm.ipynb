#### 1. Implement Bresenham's Algorithm

```python
import matplotlib.pyplot as plt
def bresenham_line(x1, y1, x2, y2):
    xes, yes = [], []
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    sx = 1 if x2 >= x1 else -1
    sy = 1 if y2 >= y1 else -1
    x, y = x1, y1
    if dx > dy:
        p = 2*dy - dx
        for _ in range(dx + 1):
            xes.append(x)
            yes.append(y)
            x += sx
            if p >= 0:
                y += sy
                p += 2*(dy - dx)
            else:
                p += 2*dy
    else:
        p = 2*dx - dy
        for _ in range(dy + 1):
            xes.append(x)
            yes.append(y)
            y += sy
            if p >= 0:
                x += sx
                p += 2*(dx - dy)
            else:
                p += 2*dx
    return xes, yes
def plot_bresenham(x1, y1, x2, y2):
    xes, yes = bresenham_line(x1, y1, x2, y2)
    plt.figure(figsize=(6, 6))
    plt.plot(xes, yes, marker='o', linestyle='-', color='green')
    plt.title("Bresenham Line Drawing Algorithm")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()
plot_bresenham(2, 3, 10, 8)
```

#### 2. Draw lines for different octants and compare visually with DDA lines.

```python
import matplotlib.pyplot as plt
def bresenham_line (x1,y1,x2,y2):
    xes, yes= [], []
    dx=abs(x2-x1)
    dy=abs(y2-y1)
    if x2>=x1:
        sx=1
    else:
        sx=-1
    if y2>=y1:
        sy=1
    else:
        sy=-1
    x,y=x1,y1
    if dx>=dy:
        p=2*dy-dx
        for i in range(dx+1):
            xes.append(x)
            yes.append(y)
            x+=sx
            if p>=0:
                y+=sy
                p+=2*dy-2*dx
            else:
                p+=2*dy
    else:
        p=2*dx-dy
        for i in range(dy+1):
            xes.append(x)
            yes.append(y)

            y+=sy
            if p>=0:
                x+=sx
                p+=2*dx-2*dy
            else:
                p+=2*dx
    return xes,yes
def dda_line(x1, y1, x2, y2):
    xes, yes = [], []
    dx = x2 -x1
    dy = y2-y1 
    steps = int(max(abs(dx), abs(dy)))
    x_inc= dx /steps
    y_inc = dy/ steps
    x = x1
    y= y1
    for i in range( steps +1):
        xes.append(round(x))
        yes.append(round(y))
        x+=x_inc
        y+=y_inc
    return (xes,yes)
def plot_dda_and_bresenham(x1,y1,x2,y2):
    plt.figure(figsize=(6,6))
    xes, yes = bresenham_line(x1,y1,x2,y2)
    plt.plot(xes, yes, marker='o', linestyle='-', color='r', label='B1')
    xes, yes = dda_line(x1,y1,x2,y2)
    plt.plot(xes, yes, marker='o', linestyle='-', color='orange', label='D1')
    xes, yes = bresenham_line(-x1,y1,-x2,y2)
    plt.plot(xes, yes, marker='o', linestyle='-', color='b', label='B2')
    xes, yes = dda_line(-x1,y1,-x2,y2)
    plt.plot(xes, yes, marker='o', linestyle='-', color='pink', label='D2')
    xes, yes = bresenham_line(-x1,-y1,-x2,-y2)
    plt.plot(xes, yes, marker='o', linestyle='-', color='y', label='B3')
    xes, yes = dda_line(-x1,-y1,-x2,-y2)
    plt.plot(xes, yes, marker='o', linestyle='-', color='black', label='D3')
    xes, yes = bresenham_line(x1,-y1,x2,-y2)
    plt.plot(xes, yes, marker='o', linestyle='-', color='g', label='B4')
    xes, yes = dda_line(x1,-y1,x2,-y2)
    plt.plot(xes, yes, marker='o', linestyle='-', color='purple', label='D4')
    plt.title("Comparison Between Bresenham and DDA Line Drawing Algorithm")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.grid(True)
    plt.show()
plot_dda_and_bresenham(0,0,18,20)
```
