#### 1. Draw a sine wave using matplotlib for ùë• from 0 to 2ùúã.

```python
import numpy as np
import matplotlib.pyplot as plt
x= np.arange(0, 5*np.pi, 0.1)
y= np.sin(x)
plt.figure(figsize=(6,2))
plt.plot(x,y, color= 'r')
plt.title("Sine Wave Visualization")
plt.xlabel("Angle (Radians)")
plt.ylabel("Amplitude")
plt.show()
```

#### 2. Write a function-based Python program to compute the factorial of a number

```python
def fact(n):
    if n==0 or n==1:
        return 1
    else: 
        return fact(n-1)*n
n= int(input("Enter a Number: "))
print(f"Factorial of {n} is: ", fact(n))
```

#### 3. Write a Python script that prints all prime numbers between 1 and 100.

```python
for i in range(2, 100):
    is_prime = True
    for j in range(2, i):
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        print(i)
```
