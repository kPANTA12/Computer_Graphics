##### 1. Implement the Midpoint Circle Drawing Algorithm.

```python
import matplotlib.pyplot as plt
def plot_circle_points(xc,yc,x,y,xes,yes):
    pts=[
        (x+xc, y+yc),
        (-x+xc, y+yc),
        (x+xc, -y+yc),
        (-x+xc, -y+yc),
        (y+xc, x+yc),
        (-y+xc, x+yc),
        (y+xc, -x+yc),
        (-y+xc, -x+yc),
    ]
    for px, py in pts:
        xes.append(px)
        yes.append(py)
def midpoint_circle(r, xc= 0, yc=0):
    x= 0
    y=r
    p=1-r
    xes, yes = [], []
    plot_circle_points(xc, yc, x,y, xes, yes)
    while x<y:
        x+=1
        if p<0:
            p = p+2*x+ 1
        else:
            y-=1
            p = p+2*(x-y)+1
        plot_circle_points(xc, yc, x, y, xes, yes)
    return xes, yes
def plot_midpoint_circle(r, xc=0, yc =0):
    xes, yes = midpoint_circle(r, xc, yc)
    plt.figure(figsize=(6, 6))
    plt.scatter (xes, yes, marker='.', color ='blue')
    plt.title("Midpoint Circle Drawing Algorithm")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.axis("Equal")
    plt.show()
plot_midpoint_circle(20, 0, 0)
```

##### 2. Draw Circle With  Different Radii  and Centres

```python
import matplotlib.pyplot as plt
def plot_circle_points(xc,yc,x,y,xes,yes):
    pts=[
        (x+xc, y+yc),
        (-x+xc, y+yc),
        (x+xc, -y+yc),
        (-x+xc, -y+yc),
        (y+xc, x+yc),
        (-y+xc, x+yc),
        (y+xc, -x+yc),
        (-y+xc, -x+yc),
    ]
    for px, py in pts:
        xes.append(px)
        yes.append(py)
def midpoint_circle(r, xc= 0, yc=0):
    x= 0
    y=r
    p=1-r
    xes, yes = [], []
    plot_circle_points(xc, yc, x,y, xes, yes)
    while x<y:
        x+=1
        if p<0:
            p = p+2*x+ 1
        else:
            y-=1
            p = p+2*(x-y)+1
        plot_circle_points(xc, yc, x, y, xes, yes)
    return xes, yes
def plot_midpoint_circle(r, xc=0, yc =0):
    plt.figure(figsize=(6, 6))
    for i, x_center, y_center in zip(r,xc, yc):
        xes, yes = midpoint_circle(i, x_center, y_center)
        plt.scatter(xes, yes, marker ='.', label= f"r = {i} ")
    plt.title("Midpoint Circle Drawing Algorithm")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.axis("Equal")
    plt.legend()
    plt.show()
plot_midpoint_circle([5, 10, 15, 20],[15, 10, 5, 0],[15, 10, 5, 0])
```
